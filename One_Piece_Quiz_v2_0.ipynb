```
#creating a list of players and player scores
players = []
player_scores = []

#Creating a Header
def header():

  print("One Piece Quiz")

  username = input("\nPlease enter your name: ").title()
  print()
  #adding of value on the list
  players.append(username)


#Quiz items
def quiz():
  header()
  score1 = 0
  ques1 = input("Which pirates is Tama waiting for?:\n(A) Buggy\n(B) Ace\n(C) Zoro\n(D) Roger\n\n")
  if ques1 == 'B':
    print("Correct")
    score1+=1
  else:
    print("Wrong")
  ques2 = input("\nWhich is the main means of communication in One Piece world?\n(A) Phone Shells\n(B) News Coos\n(C) Telegram Beetles\n(D) Transponder Snails\n\n")
  if ques2 == 'D':
    print("Correct")
    score1+=1
  else:
    print("Wrong")
  ques3 = input("\nWho created athe poisonous gas, Koro?\n(A) Avelon\n(B) Vergo\n(C) Caesar\n(D) Jack\n\n")
  if ques3 == 'C':
    print("Correct")
    score1+=1
  else:
    print("Wrong")
  ques4 = input("\nWhat color is Ichiji's raid suit?\n(A) Yellow\n(B) Red\n(C) Pink\n(D) Green\n\n")
  if ques4 == 'B':
    print("Correct")
    score1+=1
  else:
    print("Wrong")
  ques5 = input("\nHow many swords does Zoro wield when activating Toro, Ootoro and Salmon?\n(A) Three\n(B) Nine\n(C) One\n(D) Two\n\n")
  if ques5 == 'D':
    print("Correct")
    score1+=1
  else:
    print("Wrong")
  ques6 = input("\nWho does Law name as the strongest creature in the world?\n(A) Blackbeard\n(B) Big Mom\n(C) Kaido\n(D) Shanks\n\n")
  if ques6 == 'C':
    print("Correct")
    score1+=1
  else:
    print("Wrong")
  ques7 = input("\nWho was in the picture frame that made Big Mom devastated when it was broken?\n(A) Katakuri\n(B) Kaido\n(C) Mother Caramel\n(D) Lola\n\n")
  if ques7 == 'C':
    print("Correct")
    score1+=1
  else:
    print("Wrong")
  ques8 = input("\nWhat tattoo did Nami have, before the one she has currently?\n(A) Arlong's Mark\n(B) Tangerine Trees\n(C) Pinwheels\n(D) Money\n\n")
  if ques8 == 'A':
    print("Correct")
    score1+=1
  else:
    print("Wrong")
  ques9 = input("\nWhat pirate crews were Shanks and Buggy both members of when they were younger?\n(A) Cook Pirates\n(B) Whitebeard Pirates\n(C) Roger Pirates\n(D) Rumbar Pirates\n\n")
  if ques9 == 'C':
    print("Correct")
    score1+=1
  else:
    print("Wrong")
  ques10 = input("\nWhat part of the ocean is the island Amazon Lily located on?\n(A) Calm Belt\n(B) Grand Line\n(C) West Blue\n(D) North Blue\n\n")
  if ques10 == 'A':
    print("Correct\n")
    score1+=1
  else:
    print("Wrong\n")

  #This function is created to check your score
  final_score = (score1 / 10) * 100

  if final_score > 70:
    print("Congratulations! Your final score is: ", int(final_score),"%.\n\nYou are a certified One Piece Fan.\n")
  else:
    print("Your final score is: ", int(final_score),"%. Try again next time!\n\nBetter refresh and watch it again.\n")

  #adding value to the players score and initialize menu
  player_scores.append(final_score)
  menu()

#function that deletes value on the list
def remove(list1, pos):
  list1.pop(pos)

#function that can edit the value on the list
def edit_details(list2, pos, value):
  list2[pos] = value

#this function display the list of players
def player_list():
  player_info = input("Do you want to go back to the list of players?\n1. Yes\n2. No\n\n")
  if player_info == '1':
    player_information()
  else:
    menu()

#this function display the players info and scores
def player_information():
  player = 0
  print("\nList of players:\n")
  for x in range(len(players)):
    print("Player Name: ", players[x], " Score: ", player_scores[x])
  print("\n")

  #create an option where the player can choose whether to delete or edit
  option = input("Update details of players?\n(Y) Yes\n(N) No\n\n")
  if option == 'Y' or option == 'y':
    print("\nList of players:\n")
    for x in range(len(players)):
      print("(",x,") ","Player Name: ", players[x], " Score: ", player_scores[x])
    print("\n")
    
    choosing_players = int(input("Choose a player you want to update:\n"))
    player = choosing_players

    edit_delete = input("\nChoose an action:\n1.Edit\n2.Delete\n\n")
    if edit_delete == '2':
      remove(players, player)
      remove(player_scores, player)
      
      print("\nThe player has been deleted.\n")

      player_list()

    elif edit_delete == '1':
      edit_name = input("\nEnter new name:\n")
      edit_details(players, player, edit_name)

      print("\nName of player has been changed\n")

      player_list()
    
    else:
      print("Please enter a valid choice.")
      player_information()

  if option == 'N' or option == 'n':
    print("\n")
    menu()

#creating a menu for the One Piece Quiz
def menu():
  option = input("Choose 1 on the following:\n1. Play a game?\n2. See all players and scores\n3. Quit\n\n")
  if option == '1':
    print("\n")
    quiz()
  elif option == '2':
    player_information()
  else:
    print("Thank you playing the One Piece Quiz")

menu()


  

```
