```
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('insurance_regression.csv')
charges = data['charges']
age = data['age']
bmi = data['bmi']
smoker = data['smoker']

plt.bar(age,charges, color="#008fd5", label = 'age')


plt.legend()

plt.title('Insurance Charges by Age')
plt.xlabel('age')
plt.ylabel('Insurance Charges')

plt.tight_layout()

plt.show()
```
